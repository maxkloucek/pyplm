```python
h_ins = []
h_outs = []
for i in range(0, 1):
    in_mod = in_models[i]
    out_mod = out_models[i]
    h_in = np.mean(in_mod.diagonal())
    h_out = np.mean(out_mod.diagonal())
    h_ins.append(h_in)
    h_outs.append(h_out)

h_ins = np.array(h_ins)
h_outs = np.array(h_outs)
h_err = np.abs(h_outs-h_ins)
plt.plot(h_ins, h_err)
plt.show()
```
